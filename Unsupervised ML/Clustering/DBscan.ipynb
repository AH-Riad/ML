```python
from sklearn.datasets import load_iris
import seaborn as sns
from sklearn.preprocessing import StandardScaler
```

```python
iris = load_iris()

X = iris.data
y = iris.target
```

```python
scaler = StandardScaler ()
X_scaled = scaler.fit_transform(X)
```

```python
# Visualization
sns.scatterplot(
    x = X_scaled[:, 0],
    y = X_scaled[:, 2]
)
```

```python
# DBSCAN

from sklearn.cluster import DBSCAN
dbscan = DBSCAN(
    eps = 0.8,
    min_samples = 5
)
```

```python
labels = dbscan.fit_predict(X_scaled)
```

```python
# Visualization
sns.scatterplot(
    x = X_scaled[:, 0],
    y = X_scaled[:, 2],
    c = labels
)
```

## Non Linear data

```python
from sklearn.datasets import make_moons

X, y = make_moons(
    n_samples = 300,
    noise = 0.05,
    random_state = 42
)
```

```python
X_scaled = scaler.fit_transform(X)
sns.scatterplot(
    x = X[:, 0],
    y = X[:, 1]
)
```

```python
# k-means

from sklearn.cluster import KMeans

kmeans = KMeans(
    n_clusters = 2,
    random_state  = 42,
)

labels = kmeans.fit_predict(X_scaled)
```

```python
sns.scatterplot(x = X_scaled[:, 0], y = X_scaled [:, 1], c = labels )
```
