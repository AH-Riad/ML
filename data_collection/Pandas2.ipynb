```python
#which are the top 3 teams V Kohli scored highest runs ?
vk = delivery[delivery['batsman'] == 'V Kohli']
vk.groupby('bowling_team')['batsman_runs'].sum().sort_values(ascending = False).head(3)
```

```python
#Function to check any batsman highest runs against top 3 team
def run_scored(batsman_name):
  mask = delivery['batsman'] == batsman_name
  top3 =delivery[mask]
  return top3.groupby('bowling_team')['batsman_runs'].sum().sort_values(ascending = False).head(3)
```

```python
run_scored('DA Warner')
```

```python
#isin function
#Find the most destructive death over batsman in the history of IPL
#Strike Rate = (Number of runs/Number of balls)/100
#Min played in 200 balls in over 16-20
```

```python
mask = delivery['over']>15
death_over = delivery[mask]
all_batsman = death_over.groupby('batsman')['batsman_runs'].count()
x = all_batsman>200
selected_batsman = all_batsman[x].index.to_list()

```

```python
#Runs scored by all these 43 batsman
#Ball played by these 43 batsman

mask1 = delivery['batsman'].isin(selected_batsman)
final = death_over[mask1]

runs = final.groupby('batsman')['batsman_runs'].sum()
balls = final.groupby('batsman')['batsman_runs'].count()
```

```python
sr = (runs/balls)*100
sr
```
